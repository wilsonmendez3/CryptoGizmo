```python
#Import the libraries 
import numpy as np
import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
#Load the data
from google.colab import files
files.upload()
```

```python
#Store the data
df = pd.read_csv('shib-usd-max.csv')

#Set the index to the date
df = df.set_index(pd.DatetimeIndex(df['Date'].values))

#Show the data
df = df.sort_index()
df
```

```python
#Get the close price
df = df[['Close']]

#Show data
df
```

```python
#Create variable to store the number of days into the future that we want to predict
prediction_days = 1

#Create a new column called Prediction
df['Prediction'] = df[['Close']].shift(-prediction_days)

#Drop Nan
df = df.dropna()

#Show data. 
df
```

```python
#Create the independent data set (X)
X = np.array(df.drop(['Prediction'], 1))
#Remove the last n+1 row of data where n is the prediction_days
X = X[:len(df) -prediction_days -1]


#Print Data. This create a list that prints the Close two records before the final date. This will help us to validate the model later.
print(X)
```

```python
#Create dependent data set (y)
y = np.array(df['Prediction'])
#Get all of the y values except the last n+1 rows

y =  y[:-prediction_days - 1]
#Show the data
print(y)
# Y dataset
```

```python
#Split the data into 80% training data set and a 20% testing data set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
#Use the RandomForestRegressor for the model
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators = 2, random_state = 0)
forest.fit(x_train, y_train)
print(forest.score(x_test, y_test))

```

```python
#Show how close the predicted values and the actual values are
prediction = forest.predict((x_test))

#Print the rpedicted test values
print(prediction)
print()
print(y_test)
```

```python
from sklearn import metrics
r_square = metrics.r2_score(y_test, prediction)
print('R-square Error associated with Random Forest Regression is:', r_square)
```

```python
#Get the Validation data for the model
#Create a variable to store all of the rows in the data set except the last n rows
temp_df = df[:-prediction_days]

#Create a variable to store the independent price values
x_val = temp_df.tail(1)['Close'][0]

#Show the data, does is match with the 2nd to last record?
print(x_val)
```

```python
prediction = forest.predict([[x_val]])
#Print the price of Shib for next n days
print('The Price of Shib in', prediction_days, 'day(s) is predicted to be', prediction)
#print the actual value for the next n days
print('The actual price was', temp_df.tail(1)['Prediction'][0])
```

```python
import matplotlib.pyplot as plt

#Visualize the RFR by creating a range of values from min/max values of x_train
XX_val = np.arange(min(x_train), max(x_train), 0.01)
print(XX_val)
```

```python
#Reshape the data into a len(X_val)*1 array in order to make a column out of the X_Val values
XX_val = XX_val.reshape((len(XX_val), 1))

#Define a scatter plot for training data
plt.scatter(x_train, y_train, color = 'blue')

#Plot the predicted data
plt.plot(XX_val, prediction, color ='red')

#Define title, xaxis & yaxis
plt.title('Shib Price Prediction using Forest Regression')
plt.xlabel('PREDICTED USD')
plt.ylabel('ACTUAL USD')

#Set the size of the plot for better clarity
plt.figure(figsize=(1,1))

#Draw plot
plt.show()
```
